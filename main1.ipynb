```python
import os
import pandas as pd

# Path to original dataset
DATASET_DIR = "realwaste"

data = []

for class_name in os.listdir(DATASET_DIR):
    class_path = os.path.join(DATASET_DIR, class_name)
    
    if os.path.isdir(class_path):
        for img_name in os.listdir(class_path):
            data.append({
                "image_name": img_name,
                "class": class_name
            })

df = pd.DataFrame(data)
df
```

```python
df['class'].value_counts()
```

```python
df.to_csv('image class.csv')
```

```python
import shutil

NEW_IMAGE_DIR = "all_images"
os.makedirs(NEW_IMAGE_DIR, exist_ok=True)

for _, row in df.iterrows():
    src_path = os.path.join(DATASET_DIR, row["class"], row["image_name"])
    
    new_image_name = f"{row['image_name']}"
    dst_path = os.path.join(NEW_IMAGE_DIR, new_image_name)
    
    shutil.copy(src_path, dst_path)
```
